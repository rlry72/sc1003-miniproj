## Analysis of requirements

We need to sort 6000 students into groups. These 6000 students are separated into 120 tutorial groups, each with 50 students.
The sorting is done within the tutorial groups, into 10 groups of 5, so we just need to compare 50 students.

### 120 tutorial groups, 50 students/tutorial group, 10 groups of 5 students

### Train of thought
The goal is to make it as diverse as possible.
3 factors that are different among the students: Gender, School, CGPA.
Meaning: if all 3 are as different as possible, it will be the most diverse.

### Caveat
There might be some teams where diversity is just simply lacking, for example if within tutorial group there is overwhelming majority of a gender, so one team might be made up completely of one gender while the others are more diverse. 

### Workaround
We use a diversity score to compare the teams. We swap members between the most and least diverse teams (within the tutorial group) to try and balance out the teams.

```python
def diversityScore(team):
    divScore = 0
    studentIdList = list(team.keys())
    i = 0
    while i+1 < len(studentIdList):
        if team[studentIdList[i]]['Gender'] == team[studentIdList[i+1]]['Gender']:
            divScore -= 1
        if team[studentIdList[i]]['School'] == team[studentIdList[i+1]]['School']:
            divScore -= 1
        i += 1
    return divScore
```
